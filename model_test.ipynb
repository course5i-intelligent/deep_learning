```python
img = cv2.imread('data/processed/test/pituitary_tumor/image(11).jpg')
```

```python
model = load_model("models/trained.h5")
```

```python
plt.imshow(img)
```

```python
def pred(img):
    img = cv2.resize(img,(255,255))
    img = np.array(img)
    img = np.expand_dims(img, axis=0)
    pred = model.predict(img)
    pred = (str(pred))
```

```python
demo = pred(img)
```
